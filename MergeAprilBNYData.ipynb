```python
"""
Merge Gnews files to remove duplicated articles.
Date range can be configered. 
Fill summary with title if not available.
"""

import pandas as pd
from datetime import datetime, timedelta 

dir = 'key_pipeline/data/cnbc/'
df_list = []
DATE_I = datetime(2025, 8, 1)
END_DATE = datetime(2025, 9, 30)
while DATE_I != END_DATE:
    df_list.append(pd.read_csv(dir+f'{DATE_I.strftime('%Y-%m-%d')}.csv'))
    DATE_I = DATE_I + timedelta(days=1)

DATE_I = datetime(2025, 11, 16)
END_DATE = datetime(2025, 11, 30)
while DATE_I != END_DATE:
    df_list.append(pd.read_csv(dir+f'{DATE_I.strftime('%Y-%m-%d')}.csv'))
    DATE_I = DATE_I + timedelta(days=1)

DATE_I = datetime(2026, 2, 16)
END_DATE = datetime(2026, 3, 11)
while DATE_I != END_DATE:
    df_list.append(pd.read_csv(dir+f'{DATE_I.strftime('%Y-%m-%d')}.csv'))
    DATE_I = DATE_I + timedelta(days=1)

# dir = 'key_pipeline/data/reuters/'
# DATE_I = datetime(2025, 8, 1)
# END_DATE = datetime(2025, 9, 15)
# while DATE_I != END_DATE:
#     df_list.append(pd.read_csv(dir+f'{DATE_I.strftime('%Y-%m-%d')}.csv'))
#     DATE_I = DATE_I + timedelta(days=1)

df = pd.concat(df_list)
df.reset_index(inplace=True, drop=True)
print(f'df non-dedup shape: {df.shape}')
```

```python
df['title_len'] = df['title'].str.len()
df = df.loc[df.groupby('id')['title_len'].idxmax()]
df = df.reset_index(drop=True)
print(f'df dedup shape: {df.shape}')
```

```python
# save grouped csv
fname = 'key_pipeline/data/aug_mar.csv'
df.to_csv(fname, index=False)
print(f'Merged Data saved as: {fname}')
```

```python
"""
Merge Gnews April 27rd-May 22nd
"""

import pandas as pd
from datetime import datetime, timedelta 

df_list = []
START_DATE = datetime(2025, 1, 1)
for i in range(7):
    DATE_I = START_DATE + timedelta(days=i)
    df_list.append(pd.read_csv(f'data/vix_training/{DATE_I.strftime('%Y-%m-%d')}.csv'))

df_1 = pd.concat(df_list)
```

```python
df_list = []
START_DATE = datetime(2025, 1, 20)
for i in range(11):
    DATE_I = START_DATE + timedelta(days=i)
    df_list.append(pd.read_csv(f'data/vix_training/{DATE_I.strftime('%Y-%m-%d')}.csv'))

df_2 = pd.concat(df_list)
```

```python
df = pd.concat([df_1, df_2])
```

```python
df.reset_index(inplace=True, drop=True)
df['title_len'] = df['title'].str.len()
df = df.loc[df.groupby('id')['title_len'].idxmax()]
```

```python
df = df.reset_index(drop = True)
```

```python
df.to_csv('data/dxy_training/jan_partial.csv', index=False)
```
